# **Problem Statement**

**This project aims to perform an exploratory data analysis on T20 World Cup match data by combining match-level attributes with final match results. The objective is to identify key factors that contribute to winning a match, quantify their impact using statistical measures, and uncover hidden patterns that explain match outcomes.**

# **Setup**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Data Loading**

```python
df1 = pd.read_csv('/kaggle/input/t20-world-cup-2026-match-dataset/Match_dataset.csv')
df2 = pd.read_csv('/kaggle/input/t20-world-cup-2026-match-dataset/match_results.csv')
```

 # **Explore Data**

```python
df2.head()
```

```python
merged_df = pd.merge(
    df1,
    df2,
    on='Match_ID',
    how='inner'
)
```

```python
merged_df.info()
```

```python
merged_df.describe()
```

```python
# Number of matches in dataset
merged_df['Match_ID'].nunique()
```

```python
#Average match total
merged_df['Match_Total'].mean()
```

```python
# Toss Dependency
merged_df.groupby('Toss_Decision')['Winner'].count()
```

```python
# Toss winning rate

toss_win_rate = (
    merged_df['Toss_Winner'] == merged_df['Winner']
).mean() * 100

toss_win_rate
```

```python
#number of matches played by each team

matches_played = pd.concat(
    [merged_df['Team_A'],merged_df['Team_B']]
).value_counts()

matches_played
```

# **Visualisation**

```python
# Comparison between teams played most number of matches

matches_played = pd.concat(
    [merged_df['Team_A'],merged_df['Team_B']]
).value_counts()

matches_played.plot(kind = 'bar')
plt.title('matches played by teams')
plt.xlabel('team')
plt.ylabel('Number of matches')
plt.show()
```

```python
# distribution of match total scores

merged_df['Match_Total'].plot(kind = 'hist',bins = 10)

plt.title('Total scores by match')
plt.xlabel('Total Runs')
plt.ylabel('frequency')
plt.show()
```

```python
# Scores by venues

merged_df.boxplot(
    column = 'Match_Total',
    by = 'Venue',
    rot = 45
    
)

plt.title('Match totals by venues')
plt.xlabel('Venues')
plt.ylabel('Match Total')
plt.show()
```

```python
# percentage of match winners by toss winners

toss_result = (
    merged_df['Toss_Winner'] == merged_df['Winner']
).value_counts()

plt.figure()
plt.pie(
    toss_result,
    labels = ('Toss winner won','Toss winner lost'),
    autopct='%1.1f%%'
)

plt.title('Match Results based on toss')
plt.show()
    
```

# **Key Insights**

* Winning the toss provides an advantage, but it is not decisive.
* Team ranking strongly correlates with match outcomes.
* Ranking and form combined are more predictive than any single factor.
* Venue characteristics influence match dynamics beyond team strength.
* T20 matches remain inherently unpredictable despite measurable trends.
